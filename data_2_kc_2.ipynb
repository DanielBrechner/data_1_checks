```python
import pandas as pd 
```

Reading in Louisville Metro Foreclosure Data

```python
df = pd.read_csv('/Users/danielbrechner/data_1_checks/assets/Louisville_Metro_KY_-_Property_Foreclosures.csv')
```

Looking at the DataFrame to see what to clean

```python
print(df.shape)
df.head()
```

Column names are already in snake case, but lets change the column names to lowercase for consistency 

```python
df.columns = df.columns.str.lower()
df.head()
```

Finding null values

```python
null_counts = df.isnull().sum()
null_counts
```

Let's convert the zipcode to 'str' instead of 'int'. But we need to do something with NaN values first. Let's convert NaN values in the zip column to -1

```python
df['zip'] = df['zip'].fillna(-1)
```

```python
df.zip.isnull().any()
```

Now we need to convert dtype to int to drop the decimal, then convert to dtype str and only use the first five digits of each value

```python
df['zip'] = df['zip'].astype(int).astype(str).str[:5]
```

```python
df.head()
```

Let's look at the purchaser column and see if we can make some of the same values with different names consistent 

```python
unique_values = df['purchaser'].unique()
unique_values
```

```python
df['purchaser'].isnull().any()
```

We need to drop null values before cleaning

```python
df = df.dropna(subset=['purchaser'])
```

```python
df['purchaser'].isnull().any()
```

Looks like there are many different values for 'Metro', let's change them all to same value. 

```python
df.loc[df['purchaser'].str.contains('metro', case=False), 'purchaser'] = 'metro'
```

```python
df['purchaser']
```

Here we see metro values are consistent
